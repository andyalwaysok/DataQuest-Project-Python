```python
import numpy as np
top_10_major = recent_grads.loc[:10, 'Major'].values
fig = plt.figure(figsize=(20, 10)) 

barWidth = 0.5

bar1 = recent_grads['Men'].head(10)
bar2 = recent_grads['Women'].head(10)

r1 = np.arange(len(bar1))
r2 = [x + barWidth for x in r1]

plt.bar(r1, bar1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Men')
plt.bar(r2, bar2, color='#557f2d', width=barWidth, edgecolor='white', label='Women')

plt.xlabel('Majors', fontweight='bold')
plt.ylabel('Number of Men & Women', fontweight='bold')
plt.title('# of Men and Women in each category of top 10 majors')
plt.xticks([r + barWidth for r in range(len(bar1))], top_10_major, rotation=90)

plt.legend()
plt.show()
```
