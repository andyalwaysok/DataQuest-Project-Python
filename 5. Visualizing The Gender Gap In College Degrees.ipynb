Goal: Visualize the gender gap across college degrees.

Because there are seventeen degrees that we need to generate line charts for, we'll use a subplot grid layout of 6 rows by 3 columns. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

women_degrees
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
spines = ['top', 'bottom', 'left', 'right']

fig = plt.figure(figsize=(12, 18))
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)
for i in range(3,21,3):
    ax1 = fig.add_subplot(6,3,i-2)
    ax1.plot(women_degrees['Year'], women_degrees[stem_cats[(i//3)-1]], c=cb_dark_blue, linewidth=3)
    ax1.plot(women_degrees['Year'], 100-women_degrees[stem_cats[(i//3)-1]], c=cb_orange, linewidth=3)
    ax1.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax1.set_yticks([0,100])
    ax1.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if i == 3:
        ax1.text(women_degrees['Year'][women_degrees.shape[0]-1], women_degrees[stem_cats[(i//3)-1]][women_degrees.shape[0]-1], 'Women')
        ax1.text(women_degrees['Year'][women_degrees.shape[0]-1], 100-women_degrees[stem_cats[(i//3)-1]][women_degrees.shape[0]-1], 'Men')
    elif i == 18:
        ax1.text(women_degrees['Year'][women_degrees.shape[0]-1], women_degrees[stem_cats[(i//3)-1]][women_degrees.shape[0]-1], 'Women')
        ax1.text(women_degrees['Year'][women_degrees.shape[0]-1], 100-women_degrees[stem_cats[(i//3)-1]][women_degrees.shape[0]-1], 'Men')
        ax1.tick_params(labelbottom='on')
        ax1.set_xlim(1968,2011)
    for s in spines:
        ax1.spines[s].set_visible(False)
        
for j in range(2, 16, 3):
    ax2 = fig.add_subplot(6,3,j)
    ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[(j//3)]], c=cb_dark_blue, linewidth=3)
    ax2.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[(j//3)]], c=cb_orange, linewidth=3)
    ax2.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax2.set_yticks([0,100])
    ax2.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)    
    if j == 2:
        ax2.text(women_degrees['Year'][women_degrees.shape[0]-1], women_degrees[lib_arts_cats[(j//3)]][women_degrees.shape[0]-1], 'Women')
        ax2.text(women_degrees['Year'][women_degrees.shape[0]-1], 100-women_degrees[lib_arts_cats[(j//3)]][women_degrees.shape[0]-1], 'Men')        
    elif j == 14:
        ax2.tick_params(labelbottom='on')
        ax2.set_xlim(1968,2011)
    for s in spines:
        ax2.spines[s].set_visible(False)
        
for k in range(3, 21, 3):
    ax3 = fig.add_subplot(6,3,k)
    ax3.plot(women_degrees['Year'], women_degrees[other_cats[(k//3)-1]], c=cb_dark_blue, linewidth=3)
    ax3.plot(women_degrees['Year'], 100-women_degrees[other_cats[(k//3)-1]], c=cb_orange, linewidth=3)
    ax3.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax3.set_yticks([0,100])
    ax3.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if k == 3:
        ax3.text(women_degrees['Year'][women_degrees.shape[0]-1], women_degrees[other_cats[(k//3)-1]][women_degrees.shape[0]-1], 'Women')
        ax3.text(women_degrees['Year'][women_degrees.shape[0]-1], 100-women_degrees[other_cats[(k//3)-1]][women_degrees.shape[0]-1], 'Men')
    elif k == 18:
        ax3.text(women_degrees['Year'][women_degrees.shape[0]-1], women_degrees[other_cats[(k//3)-1]][women_degrees.shape[0]-1], 'Women')
        ax3.text(women_degrees['Year'][women_degrees.shape[0]-1], 100-women_degrees[other_cats[(k//3)-1]][women_degrees.shape[0]-1], 'Men')
        ax3.tick_params(labelbottom='on')
        ax3.set_xlim(1968,2011)
    for s in spines:
        ax3.spines[s].set_visible(False)
plt.show()
plt.savefig('gender_degrees.png')
```
